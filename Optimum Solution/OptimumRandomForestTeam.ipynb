```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import RandomizedSearchCV, KFold
from sklearn.ensemble import RandomForestRegressor
import pandas as pd
import numpy as np
```

```python
df_all=pd.read_csv('../ready.csv')
df_all=df_all.drop(['search_term','Unnamed: 0','product_uid','product_title','product_description','brand','product_colour',
                    'Bullet01','Bullet02','Bullet02','Bullet03','Bullet04','Bullet05','product_height',
                   'product_width','product_weight','product_depth'],axis=1) 
df_all=df_all.fillna(0)
df_all.head()
```

```python
df_train=pd.read_csv('../train.csv',encoding='latin-1')
train_size=df_train.shape[0]
X_train=df_all[:train_size].drop(['relevance'],axis=1)
y_train=df_all[:train_size]['relevance']
```

```python
rf = RandomForestRegressor(n_jobs = -1, verbose= 1,max_features= 'log2',warm_start= True,)
```

```python
params = {
            'n_estimators': np.arange(100, 3000, 100).tolist(),
            'random_state': np.arange(100, 3000, 100).tolist(),
            'min_samples_leaf': np.arange(1, 100, 10).tolist(),
            'max_depth': np.arange(1, 500, 50).tolist(),
          }

rf = RandomForestRegressor(n_jobs = -1, n_estimators = 2500, verbose= 1, random_state =2500, max_features = "log2", min_samples_leaf = 50, warm_start=True, max_depth=100 )

k_fold = KFold(3, random_state=1)

random_search = RandomizedSearchCV(rf, param_distributions=params,scoring='neg_mean_squared_error', cv=k_fold, n_iter=50)

random_search.fit(X_train, y_train)

result = random_search.cv_results_
result
```

```python
best = np.flatnonzero(result['rank_test_score'] == 1)

#Best will contain the optimised params
```

```python
optimized={}
optimized['n_estimators'] = result['param_n_estimators'].data[best][0]
optimized['random_state'] = result['param_random_state'].data[best][0]
optimized['min_samples_leaf'] = result['param_min_samples_leaf'].data[best][0]
optimized['max_depth'] = result['param_max_depth'].data[best][0]

optimized
```

```python
X_test=df_all[train_size:].drop(['relevance'],axis=1)
y_test=df_all[train_size:]['relevance']
id_test=df_all[train_size:]['id']
```

```python
rf_opt = RandomForestRegressor(n_jobs = -1, n_estimators = optimized['n_estimators'], verbose= 1,
                           random_state = optimized['random_state'], max_features = "log2",
                           min_samples_leaf = optimized['min_samples_leaf'], warm_start=True, 
                           max_depth=optimized['max_depth'] )
rf_opt.fit(X_train, y_train)
                           
y_pred = rf_opt.predict(X_test)
```

```python
pd.DataFrame({"id": id_test, "relevance": y_pred}).to_csv('submission_David.csv',index=False)
```
